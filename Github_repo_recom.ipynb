```python
data=pd.read_csv("https://query.data.world/s/zxtxz5ro2skvafkuaycnh74g5ppp2g")
data.head(1)
```

```python
data=data.drop(["Last Update Date","Number of Stars"],axis=1)
data.head()
```

```python
data["Tags"]=data["Tags"].replace(',', ' ', regex=True)
data["Description"]=data["Description"].replace(',',' ',regex=True)
```

```python
data=data.fillna(" ")
data.head()
```

```python
def combined_features(row):
    return row['Repository Name']+" "+row['Language']+" "+row['Tags']+row['Description']
data["combined_features"] = data.apply(combined_features, axis =1)
```

```python
user_repo_index=980
```

```python
def flatten(t):
    return [item for sublist in t for item in sublist]
```

```python
def get_title_from_index(index):
  if len(data[data["index"] == index]["index"])!=0:
    return data[data["index"] == index]["index"].values.tolist()
```

```python
def find_similar_repos(user_repo_index):
  similar = list(enumerate(cosine_sim[user_repo_index]))
  sorted_similar= sorted(similar, key=lambda x:x[1], reverse=True)
  print(sorted_similar)
  test_list=[]
  for i in range(len(sorted_similar)):
    test_list.append(get_title_from_index(sorted_similar[i][0]))
  res = []
  for val in test_list:
    if val is not None :
      res.append(val)
  res=flatten(res)
  return res[:15]  



  
```

```python
find_similar_repos(user_repo_index)
```

```python
import json
```

```python
df=pd.DataFrame(columns=["Username","Repository Name","Description","Language","Tags","Url"])
for i in find_similar_repos(user_repo_index):
  df=pd.DataFrame(df)
  df.loc[len(df.index)] = data.iloc[i,:6]
  

  
  result = df.to_json(orient="records")
  parsed = json.loads(result)
print(json.dumps(parsed, indent=4))
```
